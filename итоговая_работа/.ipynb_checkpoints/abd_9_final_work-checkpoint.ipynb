### 1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
import pandas as pd
```

```python
from datetime import datetime
```

```python
import numpy as np
```

```python
ratings = pd.read_csv('u.data.csv', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
ratings.head()
```

```python
ratings['datetime'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings.head()
```

```python
movie_name = ['movie_id',
              'movie_title',
              'release_date',
              'video_release_date',
              'IMDb_URL',
              'unknown',
              'Action',
              'Adventure',
              'Animation',
              'Children_s',
              'Comedy',
              'Crime',
              'Documentary',
              'Drama',
              'Fantasy',
              'Film_Noir',
              'Horror',
              'Musical',
              'Mystery',
              'Romance',
              'Sci_Fi',
              'Thriller',
              'War',
              'Western']
```

```python
movies = pd.read_csv('u.item.csv', sep = '|', encoding='ISO-8859-1', names = movie_name)
movies.head()
```

### 2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
ratings.head()
```

```python
df_top_users = ratings[['user_id', 'item_id']].groupby('user_id').count().sort_values('item_id', ascending=False)
df_top_users.head()
```

```python
print(f'Большее количество оценок поставил пользователь с user_id: {df_top_users.index[0]}')
```

### 3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

Изменять датафрейм ratings не буду, создам новый на его основе, а то жалко данные то терять, вдруг далее по заданию пригодятся :) 

```python
top_user_id = df_top_users.index[0]
```

```python
top_user_values = ratings.loc[ratings['user_id'] == top_user_id]
top_user_values.head()
```

### 4. Добавьте к датафрейму из задания 3 столбцы:

- По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
    
- Cтолбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
movies.head()
```

```python
movies_genres = movies.loc[:,'unknown' : 'Western']
movies_genres.head()
```

```python
movies_id = movies.loc[:,'movie_id']
movies_id.head()
```

```python
movies_genres_id = movies_genres.join(movies_id)
movies_genres_id.head()
```

```python
movies_genres_id.rename(columns={'movie_id': 'item_id'}, inplace=True)
movies_genres_id.head()
```

```python
top_user_movies = top_user_values.merge(movies_genres_id, how='left', on='item_id')
top_user_movies.head()
```

```python
# жанры добавлены
```

```python
ratings.head()
```

```python
ratings_pivot = ratings[['item_id','rating']].groupby('item_id').agg(['count', 'sum'])['rating']
ratings_pivot.head()
```

```python
top_user_movies_ratings = top_user_movies.merge(ratings_pivot, how='left', on='item_id')
top_user_movies_ratings.head()
```

```python
top_user_movies_ratings.rename(columns={'count': 'count_rating', 'sum': 'sum_rating'}, inplace=True)
top_user_movies_ratings.head()
```

```python
# добавил сумму и кол-во оценок
```

### 7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
```

```python
list_df = list(top_user_movies_ratings)
list_df.remove('rating')
list_df.remove('datetime')

list_df
```

```python
X = top_user_movies_ratings[list_df]
y = top_user_movies_ratings['rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Коэфф детерменизма
model.score(X_test, y_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# расчет точности
accuracy_score(y_test, lda.predict(X_test))
```
